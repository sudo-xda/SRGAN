```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('srf_2_train_results.csv')  # Replace with your actual file path

# Plot Loss_D and Loss_G
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['Loss_D'], label='Loss_D', marker='o')
plt.plot(data['Epoch'], data['Loss_G'], label='Loss_G', marker='o')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Generator and Discriminator Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plot Score_D and Score_G
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['Score_D'], label='Score_D', marker='o')
plt.plot(data['Epoch'], data['Score_G'], label='Score_G', marker='o')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.title('Generator and Discriminator Scores')
plt.legend()
plt.grid(True)
plt.show()

# Plot PSNR
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['PSNR'], label='PSNR', marker='o', color='orange')
plt.xlabel('Epoch')
plt.ylabel('PSNR (dB)')
plt.title('Peak Signal-to-Noise Ratio (PSNR)')
plt.legend()
plt.grid(True)
plt.show()

# Plot SSIM
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['SSIM'], label='SSIM', marker='o', color='green')
plt.xlabel('Epoch')
plt.ylabel('SSIM')
plt.title('Structural Similarity Index (SSIM)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
data = pd.read_csv('training_data.csv')

# Plot Loss Ratio
if 'Loss_Ratio' in data.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(data['Epoch'], data['Loss_Ratio'], label='Loss_G/Loss_D', marker='o', color='purple')
    plt.xlabel('Epoch')
    plt.ylabel('Loss Ratio')
    plt.title('Generator to Discriminator Loss Ratio')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Learning Rate
if 'Learning_Rate' in data.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(data['Epoch'], data['Learning_Rate'], label='Learning Rate', marker='o', color='blue')
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.title('Learning Rate Schedule')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Validation Metrics (PSNR and SSIM)
if 'Val_PSNR' in data.columns and 'Val_SSIM' in data.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(data['Epoch'], data['Val_PSNR'], label='Validation PSNR', marker='o', color='orange')
    plt.plot(data['Epoch'], data['Val_SSIM'], label='Validation SSIM', marker='o', color='green')
    plt.xlabel('Epoch')
    plt.ylabel('Metric Value')
    plt.title('Validation Metrics (PSNR & SSIM)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot FID (Frchet Inception Distance)
if 'FID' in data.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(data['Epoch'], data['FID'], label='FID', marker='o', color='red')
    plt.xlabel('Epoch')
    plt.ylabel('FID')
    plt.title('Frchet Inception Distance (FID)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

# Set the style
plt.style.use('dark_background')  # Using a built-in style
sns.set_theme(style="darkgrid")  # Updated seaborn styling

# Custom color scheme
colors = {
    'model1': '#00A5E3',  # Bright blue
    'model2': '#FF6B6B',  # Coral red
    'background': '#1C1C1C',  # Dark background
    'grid': '#333333'     # Dark gray
}

def setup_plot_style():
    plt.rcParams['font.size'] = 10
    plt.rcParams['axes.titlesize'] = 12
    plt.rcParams['axes.labelsize'] = 11
    plt.rcParams['grid.alpha'] = 0.3
    plt.rcParams['lines.linewidth'] = 2

def create_training_plots(data1, data2, model1_name='Hybrid', model2_name='Srgan'):
    setup_plot_style()
    
    # Create figure with GridSpec
    fig = plt.figure(figsize=(20, 24), facecolor='white')
    gs = GridSpec(4, 2, figure=fig)
    fig.suptitle('Training Metrics Comparison', fontsize=16, y=0.95)
    
    # Set all subplot backgrounds to white
    for i in range(4):
        for j in range(2):
            ax = fig.add_subplot(gs[i, j])
            ax.set_facecolor('white')
    
    # 1. Loss Plot (Generator and Discriminator)
    ax1 = fig.add_subplot(gs[0, :])
    ax1.plot(data1['Epoch'], data1['Loss_D'], label=f'{model1_name} Loss_D', 
             color=colors['model1'], marker='o', markersize=4)
    ax1.plot(data1['Epoch'], data1['Loss_G'], label=f'{model1_name} Loss_G', 
             color=colors['model1'], linestyle='--', marker='s', markersize=4)
    ax1.plot(data2['Epoch'], data2['Loss_D'], label=f'{model2_name} Loss_D', 
             color=colors['model2'], marker='o', markersize=4)
    ax1.plot(data2['Epoch'], data2['Loss_G'], label=f'{model2_name} Loss_G', 
             color=colors['model2'], linestyle='--', marker='s', markersize=4)
    ax1.set_title('Generator and Discriminator Losses', color='black')
    ax1.set_xlabel('Epoch', color='black')
    ax1.set_ylabel('Loss', color='black')
    ax1.tick_params(colors='black')
    ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # 2. Scores Plot
    ax2 = fig.add_subplot(gs[1, :])
    ax2.plot(data1['Epoch'], data1['Score_D'], label=f'{model1_name} Score_D', 
             color=colors['model1'], marker='o', markersize=4)
    ax2.plot(data1['Epoch'], data1['Score_G'], label=f'{model1_name} Score_G', 
             color=colors['model1'], linestyle='--', marker='s', markersize=4)
    ax2.plot(data2['Epoch'], data2['Score_D'], label=f'{model2_name} Score_D', 
             color=colors['model2'], marker='o', markersize=4)
    ax2.plot(data2['Epoch'], data2['Score_G'], label=f'{model2_name} Score_G', 
             color=colors['model2'], linestyle='--', marker='s', markersize=4)
    ax2.set_title('Generator and Discriminator Scores', color='black')
    ax2.set_xlabel('Epoch', color='black')
    ax2.set_ylabel('Score', color='black')
    ax2.tick_params(colors='black')
    ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # 3. Image Quality Metrics
    ax3 = fig.add_subplot(gs[2, 0])
    ax3.plot(data1['Epoch'], data1['PSNR'], label=f'{model1_name}', 
             color=colors['model1'], marker='o', markersize=4)
    ax3.plot(data2['Epoch'], data2['PSNR'], label=f'{model2_name}', 
             color=colors['model2'], marker='o', markersize=4)
    ax3.set_title('Peak Signal-to-Noise Ratio (PSNR)', color='black')
    ax3.set_xlabel('Epoch', color='black')
    ax3.set_ylabel('PSNR (dB)', color='black')
    ax3.tick_params(colors='black')
    ax3.legend()
    
    ax4 = fig.add_subplot(gs[2, 1])
    ax4.plot(data1['Epoch'], data1['SSIM'], label=f'{model1_name}', 
             color=colors['model1'], marker='o', markersize=4)
    ax4.plot(data2['Epoch'], data2['SSIM'], label=f'{model2_name}', 
             color=colors['model2'], marker='o', markersize=4)
    ax4.set_title('Structural Similarity Index (SSIM)', color='black')
    ax4.set_xlabel('Epoch', color='black')
    ax4.set_ylabel('SSIM', color='black')
    ax4.tick_params(colors='black')
    ax4.legend()
    
    # 4. Training Parameters
    ax5 = fig.add_subplot(gs[3, 0])
    ax5.plot(data1['Epoch'], data1['Loss_Ratio'], label=f'{model1_name}', 
             color=colors['model1'], marker='o', markersize=4)
    ax5.plot(data2['Epoch'], data2['Loss_Ratio'], label=f'{model2_name}', 
             color=colors['model2'], marker='o', markersize=4)
    ax5.set_title('Loss Ratio (G/D)', color='black')
    ax5.set_xlabel('Epoch', color='black')
    ax5.set_ylabel('Ratio', color='black')
    ax5.tick_params(colors='black')
    ax5.legend()
    
    ax6 = fig.add_subplot(gs[3, 1])
    ax6.plot(data1['Epoch'], data1['Learning_Rate'], label=f'{model1_name}', 
             color=colors['model1'], marker='o', markersize=4)
    ax6.plot(data2['Epoch'], data2['Learning_Rate'], label=f'{model2_name}', 
             color=colors['model2'], marker='o', markersize=4)
    ax6.set_title('Learning Rate', color='black')
    ax6.set_xlabel('Epoch', color='black')
    ax6.set_ylabel('Learning Rate', color='black')
    ax6.tick_params(colors='black')
    ax6.set_yscale('log')
    ax6.legend()
    
    # Adjust layout
    plt.tight_layout()
    return fig

# Load and process data
try:
    data1 = pd.read_csv('Flicker2K_Hybrid__srf_4x_train_results.csv')  # Replace with your first model's CSV
    data2 = pd.read_csv('Flicker2K_srgan__srf_4x_train_results.csv')  # Replace with your second model's CSV
    
    # Create plots
    fig = create_training_plots(data1, data2, 'Hybrid', 'Srgan')
    plt.show()
    
    # Optional: Save the figure
    # fig.savefig('training_comparison.png', dpi=300, bbox_inches='tight')
    
except Exception as e:
    print(f"An error occurred: {str(e)}")
```
