```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('srf_2_train_results.csv')  # Replace with your actual file path

# Plot Loss_D and Loss_G
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['Loss_D'], label='Loss_D', marker='o')
plt.plot(data['Epoch'], data['Loss_G'], label='Loss_G', marker='o')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Generator and Discriminator Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plot Score_D and Score_G
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['Score_D'], label='Score_D', marker='o')
plt.plot(data['Epoch'], data['Score_G'], label='Score_G', marker='o')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.title('Generator and Discriminator Scores')
plt.legend()
plt.grid(True)
plt.show()

# Plot PSNR
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['PSNR'], label='PSNR', marker='o', color='orange')
plt.xlabel('Epoch')
plt.ylabel('PSNR (dB)')
plt.title('Peak Signal-to-Noise Ratio (PSNR)')
plt.legend()
plt.grid(True)
plt.show()

# Plot SSIM
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['SSIM'], label='SSIM', marker='o', color='green')
plt.xlabel('Epoch')
plt.ylabel('SSIM')
plt.title('Structural Similarity Index (SSIM)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
data = pd.read_csv('training_data.csv')

# Plot Loss Ratio
if 'Loss_Ratio' in data.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(data['Epoch'], data['Loss_Ratio'], label='Loss_G/Loss_D', marker='o', color='purple')
    plt.xlabel('Epoch')
    plt.ylabel('Loss Ratio')
    plt.title('Generator to Discriminator Loss Ratio')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Learning Rate
if 'Learning_Rate' in data.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(data['Epoch'], data['Learning_Rate'], label='Learning Rate', marker='o', color='blue')
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.title('Learning Rate Schedule')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Validation Metrics (PSNR and SSIM)
if 'Val_PSNR' in data.columns and 'Val_SSIM' in data.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(data['Epoch'], data['Val_PSNR'], label='Validation PSNR', marker='o', color='orange')
    plt.plot(data['Epoch'], data['Val_SSIM'], label='Validation SSIM', marker='o', color='green')
    plt.xlabel('Epoch')
    plt.ylabel('Metric Value')
    plt.title('Validation Metrics (PSNR & SSIM)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot FID (Fréchet Inception Distance)
if 'FID' in data.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(data['Epoch'], data['FID'], label='FID', marker='o', color='red')
    plt.xlabel('Epoch')
    plt.ylabel('FID')
    plt.title('Fréchet Inception Distance (FID)')
    plt.legend()
    plt.grid(True)
    plt.show()
```
