```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('srf_2X_WBN_train_results.csv')  # Replace with your file path

# Plot Loss_D and Loss_G
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['Loss_D'], label='Loss_D', marker='o', color='red')
plt.plot(data['Epoch'], data['Loss_G'], label='Loss_G', marker='o', color='blue')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Generator and Discriminator Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plot Score_D and Score_G
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['Score_D'], label='Score_D', marker='o', color='purple')
plt.plot(data['Epoch'], data['Score_G'], label='Score_G', marker='o', color='cyan')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.title('Generator and Discriminator Scores')
plt.legend()
plt.grid(True)
plt.show()

# Plot PSNR
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['PSNR'], label='PSNR', marker='o', color='orange')
plt.xlabel('Epoch')
plt.ylabel('PSNR (dB)')
plt.title('Peak Signal-to-Noise Ratio (PSNR)')
plt.legend()
plt.grid(True)
plt.show()

# Plot SSIM
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['SSIM'], label='SSIM', marker='o', color='green')
plt.xlabel('Epoch')
plt.ylabel('SSIM')
plt.title('Structural Similarity Index (SSIM)')
plt.legend()
plt.grid(True)
plt.show()

# Plot Loss_Ratio
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['Loss_Ratio'], label='Loss Ratio', marker='o', color='brown')
plt.xlabel('Epoch')
plt.ylabel('Loss Ratio')
plt.title('Loss Ratio Over Epochs')
plt.legend()
plt.grid(True)
plt.show()

# Plot Learning Rate
plt.figure(figsize=(12, 6))
plt.plot(data['Epoch'], data['Learning_Rate'], label='Learning Rate', marker='o', color='magenta')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
